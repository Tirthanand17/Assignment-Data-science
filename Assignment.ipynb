```python
import pandas as pd

# Specify the upload path
file_path = "C:\\Users\\Dell\\Downloads\\Active Users.xlsx"

# Load the data from the specified path
data = pd.read_excel(file_path)

# Print column names
print("Column Names:", data.columns)

# Display the first few rows of the data
print(data.head())
```

```python
# Create a new DataFrame to store growth metrics
growth_metrics = pd.DataFrame(columns=['Week', 'New Users', 'Retained Users', 'Resurrected Users', 'Churned Users', 'WAU', 'Quick Ratio'])

# Assuming 'data' contains columns 'Week' and 'Device ID'
weeks = sorted(data['Week'].unique())

for i in range(1, len(weeks)):
    current_week = weeks[i]
    previous_week = weeks[i-1]
    
    current_users = set(data[data['Week'] == current_week]['Device ID'])
    previous_users = set(data[data['Week'] == previous_week]['Device ID'])
    
    new_users = current_users - previous_users
    retained_users = current_users & previous_users
    resurrected_users = current_users - retained_users
    churned_users = previous_users - current_users
    
    new_count = len(new_users)
    retained_count = len(retained_users)
    resurrected_count = len(resurrected_users)
    churned_count = len(churned_users)
    
    wau = len(current_users)
    quick_ratio = (new_count + resurrected_count) / churned_count if churned_count != 0 else 0
    
    growth_metrics = growth_metrics.append({
        'Week': current_week,
        'New Users': new_count,
        'Retained Users': retained_count,
        'Resurrected Users': resurrected_count,
        'Churned Users': churned_count,
        'WAU': wau,
        'Quick Ratio': quick_ratio
    }, ignore_index=True)

# Display the calculated growth metrics
print(growth_metrics)
```
